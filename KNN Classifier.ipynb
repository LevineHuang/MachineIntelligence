```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets.samples_generator import make_classification

random_state = 3
# X为样本特征，Y为样本类别输出， 共1000个样本，每个样本2个特征，输出有3个类别，没有冗余特征，每个类别一个簇
# X, Y = make_classification(n_samples=1000,n_features=2,n_classes=3,n_redundant=0,n_clusters_per_class=1)
X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, n_classes=3,random_state=random_state)

plt.scatter(X[:,0], X[:,1], c=Y, marker='o', edgecolors='k', s=20)
plt.show()
```

```python
from sklearn import neighbors
n_neighbors = 15
weights = 'distance'
clf = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors,algorithm='auto',weights=weights)
clf.fit(X,Y)
# 获取当前设置的参数，可对其它参数进行相应设置
print(clf.get_params)
clf.score
```

```python
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

x1_min, x1_max = np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1
x2_min, x2_max = np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1 

x1, x2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02), np.arange(x2_min, x2_max, 0.02))
x = np.c_[x1.ravel(), x2.ravel()]

y_pred = clf.predict(x)

print("x1.shape: ", x1.shape)
print("x.shape: ", x.shape)
print("y_pred: ", type(y_pred), y_pred.shape)

y_pred = y_pred.reshape(x1.shape)

# 可视化
# plt.figure(figsize=(7,5))
# 预测边界面
plt.pcolormesh(x1, x2, y_pred, cmap=cmap_light)

# 训练数据
plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', edgecolors='k', s=20, cmap=cmap_bold)
plt.show()
```

比较不同的临近权weights

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.colors import ListedColormap

from sklearn.datasets.samples_generator import make_classification
from sklearn.neighbors import KNeighborsClassifier
```

```python
n_neighbors = 15
# weights = ['distance', 'uniform']
step = .02 # mesh step
random_state = 6

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# 生成数据
X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=random_state)

plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', edgecolors='k', s=30)
plt.show()
```

```python
# 对比不同的邻近权weights的影响

for weights in ['distance', 'uniform']:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, algorithm='auto')
    clf.fit(X, Y)
    
    # 预测数据
    x1_min, x1_max = np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1
    x2_min, x2_max = np.min(X[:, 1]) - 1, np.max(X[:, 1]) + 1
    
    x1, x2 = np.meshgrid(np.arange(x1_min, x1_max, step), np.arange(x2_min, x2_max, step))
    x = np.c_[x1.ravel(), x2.ravel()]
    
    # 预测
    y_pred = clf.predict(x)
    y_pred = y_pred.reshape(x1.shape)
    
    # 可视化, 需先画边界面，再画散点图， 否则散点图会被覆盖
    plt.figure()
    # 预测边界面
    plt.pcolormesh(x1, x2, y_pred, cmap=cmap_light)
    # 原始数据
    plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', edgecolors='k', s=30, cmap=cmap_bold)
    
    plt.title("n_neighbors = %i, weights = '%s'" % (n_neighbors, weights))
plt.show()
```
